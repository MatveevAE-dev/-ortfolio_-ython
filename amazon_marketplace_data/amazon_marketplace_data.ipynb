# Приведение столбцов к стилю camel_case

```python
def camel_case(text):
    if not text:
        return text
    first, *rest = text.replace(" ", "_").split("_")
    return ''.join([first.lower(), *map(str.capitalize, rest)])


# Применяем функцию к именам столбцов
df.columns = [camel_case(col) for col in df.columns]

# Выводим обновленные имена столбцов для проверки
print(df.columns)

df.head()
```

### Анализ клиентов по Регионам

#### Продажи и количесто клиентов

```python
region_cntCustomers_totalSales=df.groupby(['region'])[['customerId','sales']].agg(uniqueCustomers=('customerId','nunique'),totalSales=('sales','sum'))\
.sort_values(by='totalSales',ascending=False).reset_index()
# Округление totalSales до целых чисел
region_cntCustomers_totalSales['totalSales'] = region_cntCustomers_totalSales['totalSales'].round()
region_cntCustomers_totalSales['avgsales_per_customer']=region_cntCustomers_totalSales['totalSales']/region_cntCustomers_totalSales['uniqueCustomers']
region_cntCustomers_totalSales['avgsales_per_customer']=region_cntCustomers_totalSales['avgsales_per_customer'].round()
region_cntCustomers_totalSales
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.colors as pc

# Создание графика с двумя осями
fig = make_subplots(specs=[[{'secondary_y': True}]])

# Добавление столбцов для суммы продаж (левая ось)
fig.add_trace(
    go.Bar(
        x=region_cntCustomers_totalSales['region'],
        y=region_cntCustomers_totalSales['totalSales'],
        name='Сумма продаж',
        marker_color='forestgreen',  # Темно-зеленый цвет
        offsetgroup=0,
        text=region_cntCustomers_totalSales['totalSales'],
        textposition='outside',
    ),
    secondary_y=False
)

# Добавление столбцов для количества уникальных клиентов (правая ось)
fig.add_trace(
    go.Bar(
        x=region_cntCustomers_totalSales['region'],
        y=region_cntCustomers_totalSales['uniqueCustomers'],
        name='Количество клиентов',
        marker_color='mediumseagreen',  # Светло-зеленый цвет
         offsetgroup=1,
        text=region_cntCustomers_totalSales['uniqueCustomers'],
        textposition='outside',
    ),
    secondary_y=True
)

# Настройка осей и заголовка
fig.update_layout(
    title='Продажи и количество клиентов по регионам',
    xaxis_title='Регион',
    yaxis_title='Продажи',
    yaxis2_title='Клиенты',
    barmode='group',
     template='plotly_white',
    plot_bgcolor='white', # Убираем серый фон графика
    paper_bgcolor='white', # Убираем серый фон всего графика
    xaxis=dict(showgrid=False),
    yaxis=dict(showgrid=False),
    yaxis2=dict(showgrid=False),
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=-0.25,
        xanchor="center",
        x=0.5
    ),
)
# Регулируем размер графика
fig.update_layout(
    width=1200,  # Ширина графика
    height=600   # Высота графика
)

fig.show()
```

```python
# Вычисление процентов от общей суммы продаж
total_sales = region_cntCustomers_totalSales['totalSales'].sum()
region_cntCustomers_totalSales['sales_percentage'] = (region_cntCustomers_totalSales['totalSales'] / total_sales) * 100


# Цветовая палитра оттенков зеленого
colors = ['forestgreen', 'seagreen', 'mediumseagreen', 'lightgreen']

# Создание круговой диаграммы (бублика)
fig = go.Figure(
    data=[go.Pie(
        labels=region_cntCustomers_totalSales['region'],
        values=region_cntCustomers_totalSales['sales_percentage'],
        hole=.3,
        marker_colors=colors,
        textinfo='percent',
        textposition='inside',
        rotation=90
    )]
)


# Настройка заголовка
fig.update_layout(
    title='Деление продаж по регионам (%)',
    template='plotly_white',
    plot_bgcolor='white',
    paper_bgcolor='white',
    width=600,
    height=600,
    showlegend=True,
    legend=dict(
        orientation="h",  # Горизонтальное расположение
        yanchor="bottom",   # Привязка по нижней границе
        y=-0.2,           # Размещение ниже графика, можно настроить по вкусу
        xanchor="center",  # Центрирование по горизонтали
        x=0.5             # Центрирование по горизонтали
    )
)


fig.show()
```

Наибольшие продажи и количество клиентов в регионе  West . В регионе  Central заметно ниже продажи на 1-го клиента (453 у.е.) по сравнению с другими регионами.

#### Средний чек

```python
region_avgСheck=df.groupby('region')[['orderId','sales']].agg(nOrders=('orderId','nunique'),avgCheck=('sales','mean'))\
.sort_values(by='nOrders', ascending=False).reset_index()
region_avgСheck['avgCheck']=region_avgСheck['avgCheck'].round(0)
region_avgСheck
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Создаем палитру зеленых оттенков
green_palette = sns.color_palette("Greens", n_colors=len(region_avgСheck))

# Создаем фигуру и оси
fig, ax1 = plt.subplots(figsize=(12, 6))

# Столбцы для nOrders (левая ось)
bars = sns.barplot(x='region', y='nOrders', data=region_avgСheck, palette=green_palette, ax=ax1)
ax1.set_ylabel('Количество заказов', fontsize=12)
ax1.tick_params(axis='y')
ax1.grid(False)

# Создаем вторую ось (правая ось)
ax2 = ax1.twinx()

# Линия для avgCheck (правая ось)
sns.lineplot(x='region', y='avgCheck', data=region_avgСheck, color='black', marker='o', ax=ax2, linewidth=2)
ax2.set_ylabel('Средний чек', fontsize=12)
ax2.tick_params(axis='y')

# Добавление меток значений
for index, row in region_avgСheck.iterrows():
    ax1.text(row.name, row['nOrders'], int(row['nOrders']), color='black', ha='center', va='bottom')
    ax2.text(row.name, row['avgCheck'], int(row['avgCheck']), color='black', ha='center', va='bottom')


# Общий заголовок и метки оси x
plt.title('Распределение количества заказов и среднего чека по регионам', fontsize=14)
ax1.set_xlabel('Регион', fontsize=12)

# Улучшаем расположение и отображение легенды
fig.tight_layout(rect=[0, 0.05, 1, 0.90])  # добавлено rect для резервирования места под легенду


# Создаем список элементов легенды для столбцов (nOrders)
legend_elements = []
for bar, region in zip(bars.patches, region_avgСheck['region']):
    legend_elements.append((bar, region))


# Создаем элемент легенды для линии
line2, = ax2.plot([], [], color='black', marker='o', linestyle='-')
legend_elements.append((line2, 'Средний чек'))


# Добавляем легенду с наименованием регионов
handles = [elem[0] for elem in legend_elements]
labels = [elem[1] for elem in legend_elements]

fig.legend(handles, labels, loc="lower center", ncol=5)


plt.show()
```

В регионе Central наблюдается самый низки средний чек.

## Динамика заказов и продаж

```python
df.columns
```

```python
Orders=df.groupby(pd.Grouper(key='orderDate', freq='ME'))[['orderId','sales']].agg(cntOrders=('orderId','nunique'),sumSales=('sales','sum')).sort_values(by='orderDate').reset_index()
Orders['sumSales']=Orders['sumSales'].round(0)
# 3. Форматирование orderDate в название месяца
Orders['orderDate'] = Orders['orderDate'].dt.strftime('%B')
Orders
```

```python
import plotly.graph_objects as go
import numpy as np
from plotly.subplots import make_subplots

# Создание фигуры с двумя осями
fig = make_subplots(specs=[[{'secondary_y': True}]])

# Палитра зеленых оттенков
green_palette = ['#B7E4C7', '#74C69D', '#2D6A4F']

# === График для суммы продаж (левая ось) ===
x_values = np.arange(len(Orders))
y_sales = Orders['sumSales'].values

# Линейная регрессия (тренд)
trend_sales = np.polyfit(x_values, y_sales, 1)
trend_line_sales = np.poly1d(trend_sales)

fig.add_trace(go.Scatter(
    x=Orders['orderDate'],
    y=Orders['sumSales'],
    name='Сумма продаж',
    line=dict(color=green_palette[2], width=2),
    mode='lines+markers+text',
    marker=dict(size=8, symbol='circle', color=green_palette[2]),
    text=[str(sales) for sales in Orders['sumSales']],  # Метки на каждой точке
    textposition=['top center' if i % 2 == 0 else 'bottom center' for i in range(len(Orders))],
    hovertemplate='<b>%{x}</b><br>Продажи: %{y}',
), secondary_y=False)

# Линия тренда для продаж
fig.add_trace(go.Scatter(
    x=Orders['orderDate'],
    y=trend_line_sales(x_values),
    name="Тренд продаж",
    line=dict(color=green_palette[2], width=2, dash="dash"),
    mode='lines',
    hoverinfo='skip'
), secondary_y=False)

# === График для количества заказов (правая ось) ===
y_orders = Orders['cntOrders'].values
trend_orders = np.polyfit(x_values, y_orders, 1)
trend_line_orders = np.poly1d(trend_orders)

fig.add_trace(go.Scatter(
    x=Orders['orderDate'],
    y=Orders['cntOrders'],
    name='Количество заказов',
    line=dict(color=green_palette[1], width=2),
    mode='lines+markers+text',
    marker=dict(size=8, symbol='circle', color=green_palette[1]),
    text=[str(cnt) for cnt in Orders['cntOrders']],  # Метки на каждой точке
    textposition=['top center' if i % 2 == 0 else 'bottom center' for i in range(len(Orders))],
    hovertemplate='<b>%{x}</b><br>Заказы: %{y}',
), secondary_y=True)

# Линия тренда для заказов
fig.add_trace(go.Scatter(
    x=Orders['orderDate'],
    y=trend_line_orders(x_values),
    name="Тренд заказов",
    line=dict(color=green_palette[1], width=2, dash="dash"),
    mode='lines',
    hoverinfo='skip'
), secondary_y=True)

# === Настройка осей ===
fig.update_xaxes(title_text='Месяц', showgrid=True, showline=True, linewidth=1, linecolor='black')
fig.update_yaxes(title_text='Продажи', secondary_y=False, showgrid=True, showline=True, linewidth=1, linecolor='black')
fig.update_yaxes(title_text='Количество заказов', secondary_y=True, showgrid=True, showline=True, linewidth=1, linecolor='black')

# === Настройка внешнего вида ===
fig.update_layout(
    title='Продажи и количество заказов по месяцам',
    width=1200,  # Ширина графика
    height=600,  # Высота графика
    plot_bgcolor='white',
    hovermode='x unified',
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    ),
    font=dict(
        family="Arial",
        size=12,
        color="black"
    ),
    margin=dict(b=100, r=100, l=100)
)

fig.show()
```

К концу года прослеживается положительная динамика по заказам и продажам.

```python
import plotly.graph_objects as go
import numpy as np

# Создание фигуры
fig = go.Figure()

# Палитра зеленых оттенков
green_palette = ['#B7E4C7', '#74C69D', '#2D6A4F']

# Получаем список категорий
categories = OrdersCategory['category'].unique()

# Добавляем график для каждой категории
for i, category in enumerate(categories):
    df_category = OrdersCategory[OrdersCategory['category'] == category]

    # Определяем шаг для подписей (примерно 10 меток на всю длину)
    N = max(1, len(df_category) // 10)

    # Создаем массив меток и позиций
    text_labels = []
    text_positions = []

    prev_value = None  # Для контроля разницы между соседними точками
    for idx, cnt in enumerate(df_category['cntOrders']):
        # Оставляем подпись только для каждой N-й точки или если разница значительная
        if idx % N == 0 or (prev_value is not None and abs(cnt - prev_value) > 10):
            text_labels.append(str(cnt))
            # Чередуем позиции текста: вверх и вниз
            text_positions.append('top center' if idx % 2 == 0 else 'bottom center')
            prev_value = cnt
        else:
            text_labels.append("")
            text_positions.append("")

    # Добавляем сам график с данными
    fig.add_trace(go.Scatter(
        x=df_category['orderDate'],
        y=df_category['cntOrders'],
        name=category,
        line=dict(
            color=green_palette[i % len(green_palette)],  # Циклический выбор цвета
            width=2
        ),
        mode='lines+markers+text',
        marker=dict(
            size=8,
            symbol='circle',
            color=green_palette[i % len(green_palette)]
        ),
        text=text_labels,  # Добавление подписей только на нужные точки
        textposition=text_positions,  # Чередуем метки сверху и снизу
        hovertemplate='<b>%{x}</b><br>Заказы: %{y}',
        connectgaps=True
    ))

    # === Линия тренда ===
    x_values = np.arange(len(df_category))  # Числовые значения вместо дат
    y_values = df_category['cntOrders'].values

    # Линейная регрессия (аппроксимация линейной функцией)
    trend = np.polyfit(x_values, y_values, 1)  # Коэффициенты [a, b] для y = ax + b
    trend_line = np.poly1d(trend)  # Генерация уравнения тренда

    # Добавляем трендовую линию
    fig.add_trace(go.Scatter(
        x=df_category['orderDate'],
        y=trend_line(x_values),
        name=f"Тренд {category}",
        line=dict(
            color=green_palette[i % len(green_palette)],
            width=2,
            dash="dash"  # Пунктирная линия
        ),
        mode='lines',
        hoverinfo='skip'
    ))

# Настройка внешнего вида
fig.update_layout(
    title='Динамика количества заказов по категориям',
    xaxis_title='Месяц',
    yaxis_title='Количество заказов',
    plot_bgcolor='white',
    hovermode='x unified',
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    ),
    font=dict(
        family="Arial",
        size=12,
        color="black"
    ),
    width=1200,  # Ширина графика
    height=600   # Высота графика
)

# Настройка осей (добавление линий)
fig.update_xaxes(showgrid=True, showline=True, linewidth=1, linecolor='black')
fig.update_yaxes(showgrid=True, showline=True, linewidth=1, linecolor='black')

# Форматирование чисел
fig.update_layout(
    yaxis=dict(tickformat=",")
)

fig.show()
```

К концу года прослеживается положительный тренд по всем категориям. При этом можно выделить всплеск заказов в сентябре и ноябре в каждой из категорий.
